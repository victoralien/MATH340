### 2.b

```python
# Question 2
# Add your code for Question 2 here (inside a code cell)
import numpy as np

def softmaxLoop(x):
    output = np.zeros_like(x,dtype = float)
    expValues = np.zeros_like(x,dtype = float)

    for i in range(len(x)):
        expValues[i] = np.exp(x[i])

    sumExp = 0
    for i in range(len(expValues)):
        sumExp += expValues[i]

    for i in range(len(x)):
        output[i] = expValues[i] / sumExp

    return np.array(output)


def softmax(x):
    exp_x = np.exp(x)
    return exp_x / exp_x.sum()

input_vector = np.array([10,2,.1])

output_vector = softmaxLoop(input_vector)

output = softmax(input_vector)

print(f"Input vector: {input_vector}\nSoftmax (looped) output: {output_vector}\nSum of Softmax (looped) Output: {output_vector.sum()}")
print(f"\nInput vector: {input_vector}\nSoftmax output: {output}\nSum of Softmax Output: {output.sum()}")

# Run the cell to display your results
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

t = np.linspace(0, 3, 300)

sigma_values = [0.1, 0.2, 0.3, 0.4]

max_y = np.exp(sigma_values[-1] * t[-1])  

for i, sigma in enumerate(sigma_values):
    y = np.exp(sigma * t)
    
    axes[i].plot(t, y, 'b-', linewidth=2)
    
    axes[i].set_xlabel('t')
    axes[i].set_ylabel('y')
    axes[i].set_title(f' = {sigma}')
    
    axes[i].set_xlim(0, 3)
    
    axes[i].set_ylim(0, max_y * 1.05)
    
    axes[i].grid(True, linestyle='--', alpha=0.7)

plt.suptitle('Solutions of y = e^(*t) for Different  Values', fontsize=16)

plt.tight_layout()
plt.subplots_adjust(top=0.9) 

plt.show()
```
