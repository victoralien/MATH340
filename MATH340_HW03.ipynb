### Results for Question 2 (within a Markdown cell)

```python
import numpy as np
import matplotlib.pyplot as plt
# Question 2
# Add your code for Question 2 here (inside a code cell)
def relu(x):
    return np.maximum(0, x)

def logistic(x):
    return 1 / (1 + np.exp(-x))

def hyperbolicTangent(x):
    return np.tanh(x)

x = np.linspace(-1.5, 1.5, 100)

logisticValues = logistic(x)
hyperbolicTangentValues = hyperbolicTangent(x)
reluValues = relu(x)


plt.figure(figsize=(10, 5))
plt.plot(x, logisticValues, label='Logistic')
plt.plot(x, hyperbolicTangentValues, label='Hyperbolic Tangent')
plt.plot(x, reluValues, label='ReLU')  


plt.legend()
plt.title('Activation Functions')
plt.grid(True)
plt.show()

# Run the cell to display your results
```

### Results for Question 5

```python
import math
inc = 3

for n in range(1 , 60, inc):
    r = 2
    for i in range(n):
        r = math.sqrt(r)
    for i in range(n):
        r **= 2
    print("%d times sqrt and **2: %.16f"%(n, r))
```

The code is a nested for loop. The first loop loops from 1 - 60, incremeting by 3, so the higest it will get to is 58. Inside the loop, a value r is set to 2, and then whatever the current value of the loop is, 1, 4, 7, 10, ect, the square root of r is taken that many times, then r is squared the same amout of times. Then r is printed to the console. 

Theoretically the result should always print 2, because we squared the number the same amout of times that we took it square root, but due to floating-point precision limitations, the results will not be exactly the same, which is the reasion we had to print so many decimals. As the number of square roots increases, the final result deviates further from the expected value of 2.
